```python
import pandas as pd
import requests
import json
from sqlalchemy import create_engine
```

```python
file_path = 'netflix_titles_nov_2019.csv'
```

```python
netflix_df = pd.read_csv(file_path)
```

```python
netflix_df.dropna(inplace=True)
```

```python
movie_df = netflix_df.loc[netflix_df["type"] == "Movie"]
```

```python
movie_df
```

```python
movie_titles = movie_df["title"].tolist()

print(movie_titles)
```

```python

api_key = "&apikey=trilogy"

url = "https://www.omdbapi.com/?t="

response = []
for title in movie_titles[:1000]:
    #print(url + title + api_key)
#     print(f"Collecting data on title.....{title}")
    data = requests.get(url + title + api_key)
    # Save JSON
    response.append(data.json())
```
