```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import Session
```

```python
db_path = "sqlite:///database/movies.sql"
engine = create_engine(db_path)
```

```python
session = Session(bind=engine)
sql_query = "select imdbID, Title, Year, Runtime, Genre, Rated, imdbVotes, imdbRating from test"
movies = engine.execute(sql_query).fetchall()
session.close()
```

```python
movies_lst = []

for row in movies:
    
    try:
        movies_lst.append({"id":row[0],
                       "title":row[1],
                       "year":int(row[2]),
                       "Runtime":int(row[3].strip().lower().replace(" min",""))
                      })
    except:
        pass
    
    

movies_lst
```

```python
movies2_lst = []

for row in movies:
    
    try:
        movies2_lst.append({"id":row[0],
                       "title":row[1],
                       "imdbVotes":int(row[6].replace(",","")),
                       "imdbRating":float(row[7])
                      })
    except:
        pass
    
    

movies2_lst
```
